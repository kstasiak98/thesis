```python
import tkinter as tk
import xml.etree.ElementTree as ET

def load_instructions(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    instructions = []
    roles = set()

    for instruction in root.findall('instruction'):
        content = instruction.find('content').text
        role_elems = instruction.find('roles').findall('role')
        instruction_roles = [role.text for role in role_elems]

        instructions.append((content, instruction_roles))

        for role in instruction_roles:
            roles.add(role)

    return instructions, roles

def create_gui(instructions, roles):
    root = tk.Tk()

    for idx, instruction in enumerate(instructions):
        content, instruction_roles = instruction

        # label = tk.Label(root, text=f"Instruction {idx+1}: {content} - Roles: {', '.join(instruction_roles)}")
        label = tk.Label(root, text=f"Instruction {idx+1}: {content}")
        label.pack()

    for role in roles:
        button = tk.Button(root, text=role)
        button.pack()

    root.mainloop()

def main():
    instructions, roles = load_instructions('instructions.xml')
    create_gui(instructions, roles)

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
import xml.etree.ElementTree as ET

class InstructionApplication:
    def __init__(self, root, instructions):
        self.root = root
        self.instructions = instructions
        self.roles = self.get_roles()
        self.labels = []
        self.buttons = []

        self.create_buttons()
        self.create_labels()

    def get_roles(self):
        roles = set()
        for instruction in self.instructions:
            _, instruction_roles = instruction
            for role in instruction_roles:
                roles.add(role)

        return roles

    def create_labels(self):
        for idx, instruction in enumerate(self.instructions):
            content, _ = instruction
            label = tk.Label(self.root, text=f"Instruction {idx+1}: {content}")
            label.pack()
            self.labels.append(label)

    def create_buttons(self):
        for i,role in enumerate(self.roles):
            button = tk.Button(self.root, text=role, command=lambda role=role: self.toggle_role(role))
            button.grid(row=0, column=i)
            button.pack()
            self.buttons.append(button)

    def toggle_role(self, role):
        for label, instruction in zip(self.labels, self.instructions):
            _, instruction_roles = instruction
            if role in instruction_roles:
                # The role matches the instruction, so the label should be visible
                label.config(fg='black')
            else:
                # The role does not match the instruction, so hide the label
                label.config(fg='gray')

def load_instructions(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    instructions = []

    for instruction in root.findall('instruction'):
        content = instruction.find('content').text
        role_elems = instruction.find('roles').findall('role')
        instruction_roles = [role.text for role in role_elems]

        instructions.append((content, instruction_roles))

    return instructions

def main():
    instructions = load_instructions('instructions.xml')

    root = tk.Tk()
    app = InstructionApplication(root, instructions)
    root.mainloop()

if __name__ == "__main__":
    main()
```
